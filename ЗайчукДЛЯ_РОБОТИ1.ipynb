```python
""" –ü—Ä–∏–≤—ñ—Ç!
–Ø–∫ –≤ Python –Ω–∞ Google Collab –Ω–∞–ø–∏—Å–∞—Ç–∏ –∫–æ–¥ —â–æ–± —Å—Ç–≤–æ—Ä–∏—Ç–∏ —Å–ø–∏—Å–æ–∫ Lv_49(A,B,C,D,E) –¥–æ Lv_60(A,B,C,D,E)

–ü—Ä–∏–∫–ª–∞–¥:
Lv_49A
Lv_49B
Lv_49C
Lv_49D
Lv_49E
Lv_50A
Lv_50B
Lv_50C
Lv_50D
Lv_50E
Lv_51A
Lv_51B
Lv_51C
Lv_51D
Lv_51E"""
```

```python
# –°—Ç–≤–æ—Ä—é—î–º–æ –ø–æ—Ä–æ–∂–Ω—ñ–π —Å–ø–∏—Å–æ–∫
levels = []

# –¶–∏–∫–ª –ø–æ –Ω–æ–º–µ—Ä–∞—Ö –≤—ñ–¥ 49 –¥–æ 61 –≤–∫–ª—é—á–Ω–æ
for num in range(49,61):
    # –¶–∏–∫–ª –ø–æ –ª—ñ—Ç–µ—Ä–∞—Ö A-E
    for letter in['A','B','C','D','E']:
      levels.append(f"Lv_{num}{letter}")

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
for item in levels:
  print(item)

# –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å—Ç–≤–æ—Ä–µ–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∑–Ω–∞—á–µ–Ω—å:", len(levels))
```

```python
# –°—Ç–≤–æ—Ä—é—î–º–æ –ø–æ—Ä–æ–∂–Ω—ñ–π —Å–ø–∏—Å–æ–∫
levels_ = []

# Lv_Vitr_SKF —â–µ Lv_Vitr_SKF(–ê,–í,–°,D) —Ç—É—Ç –º–∞—î —Å—Ç–≤–æ—Ä–∏—Ç–∏—Å—è 4 –ø–æ–ª–∏—Ü—ñ
# –¶–∏–∫–ª –ø–æ –ª—ñ—Ç–µ—Ä–∞—Ö A‚ÄìD
for letter in ['A', 'B', 'C', 'D']:
        levels_.append(f"Lv_Vitr_SKF{num}{letter}")

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
for item in levels_:
    print(item)

# –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å—Ç–≤–æ—Ä–µ–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∑–Ω–∞—á–µ–Ω—å:", len(levels_))
```

```python
def generate_list(prefix, letters, start=None, end=None):
    """
    prefix - –ø–æ—á–∞—Ç–æ–∫ –Ω–∞–∑–≤–∏ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥ 'Lv_' –∞–±–æ 'Lv_Vitr_SKF')
    letters - —Ä—è–¥–æ–∫ –∞–±–æ —Å–ø–∏—Å–æ–∫ –∑ –ª—ñ—Ç–µ—Ä–∞–º–∏ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥ 'ABCDE' –∞–±–æ ['A','B','C','D'])
    start, end - —è–∫—â–æ –ø–æ—Ç—Ä—ñ–±–Ω—ñ —á–∏—Å–ª–∞ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥ 49 –¥–æ 60
    """
    result = []

    if start is not None and end is not None:
        for num in range(start, end + 1):
            for letter in letters:
                result.append(f"{prefix}{num}{letter}")
    else:
        for letter in letters:
            result.append(f"{prefix}{letter}")

    return result


# –ü—Ä–∏–∫–ª–∞–¥ 1: Lv_49(A‚ÄìE) –¥–æ Lv_60(A‚ÄìE)
list1 = generate_list("Lv_", "ABCDE", 49, 60)
print("\n".join(list1))
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å:", len(list1))

print("-" * 30)

# –ü—Ä–∏–∫–ª–∞–¥ 2: Lv_Vitr_SKF(A‚ÄìD)
list2 = generate_list("Lv_Vitr_SKF", "ABCD")
print("\n".join(list2))
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å:", len(list2))
```

```python
import pandas as pd
from google.colab import files

# 1. –ó–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ —Ñ–∞–π–ª —É Colab
uploaded = files.upload()  # –æ–±–∏—Ä–∞—î—à ids.csv –∑—ñ —Å–≤–æ–≥–æ –∫–æ–º–ø‚Äô—é—Ç–µ—Ä–∞

# 2. –ß–∏—Ç–∞—î–º–æ CSV
df = pd.read_csv("ids.csv", header=None, dtype=str)  # –±–µ–∑ –∑–∞–≥–æ–ª–æ–≤–∫—ñ–≤, —è–∫ —Ä—è–¥–∫–∏
ids = df[0].tolist()

# 3. –î–æ–¥–∞—î–º–æ –ª–∞–ø–∫–∏
formatted_ids = [f'"{i}"' for i in ids]

# 4. –ó–∞–ø–∏—Å—É—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É txt
output_file = "formatted_ids.txt"
with open(output_file, "w") as f:
    f.write(",\n".join(formatted_ids))

# 5. –ó–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ –≥–æ—Ç–æ–≤–∏–π —Ñ–∞–π–ª –Ω–∞ –∫–æ–º–ø‚Äô—é—Ç–µ—Ä
files.download(output_file)
```
