```python
""" Привіт!
Як в Python на Google Collab написати код щоб створити список Lv_49(A,B,C,D,E) до Lv_60(A,B,C,D,E)

Приклад:
Lv_49A
Lv_49B
Lv_49C
Lv_49D
Lv_49E
Lv_50A
Lv_50B
Lv_50C
Lv_50D
Lv_50E
Lv_51A
Lv_51B
Lv_51C
Lv_51D
Lv_51E"""
```

```python
# Створюємо порожній список
levels = []

# Цикл по номерах від 49 до 61 включно
for num in range(49,61):
    # Цикл по літерах A-E
    for letter in['A','B','C','D','E']:
      levels.append(f"Lv_{num}{letter}")

# Виводимо результат
for item in levels:
  print(item)

# Порахувати кількість створених елементів
print("Кількість значень:", len(levels))
```

```python
# Створюємо порожній список
levels_ = []

# Lv_Vitr_SKF ще Lv_Vitr_SKF(А,В,С,D) тут має створитися 4 полиці
# Цикл по літерах A–D
for letter in ['A', 'B', 'C', 'D']:
        levels_.append(f"Lv_Vitr_SKF{num}{letter}")

# Виводимо результат
for item in levels_:
    print(item)

# Порахувати кількість створених елементів
print("Кількість значень:", len(levels_))
```

```python
def generate_list(prefix, letters, start=None, end=None):
    """
    prefix - початок назви (наприклад 'Lv_' або 'Lv_Vitr_SKF')
    letters - рядок або список з літерами (наприклад 'ABCDE' або ['A','B','C','D'])
    start, end - якщо потрібні числа в діапазоні, наприклад 49 до 60
    """
    result = []

    if start is not None and end is not None:
        for num in range(start, end + 1):
            for letter in letters:
                result.append(f"{prefix}{num}{letter}")
    else:
        for letter in letters:
            result.append(f"{prefix}{letter}")

    return result


# Приклад 1: Lv_49(A–E) до Lv_60(A–E)
list1 = generate_list("Lv_", "ABCDE", 49, 60)
print("\n".join(list1))
print("Кількість:", len(list1))

print("-" * 30)

# Приклад 2: Lv_Vitr_SKF(A–D)
list2 = generate_list("Lv_Vitr_SKF", "ABCD")
print("\n".join(list2))
print("Кількість:", len(list2))
```

```python
Розбір по кроках
1️⃣ Що таке def generate_list(...):
def — ключове слово, щоб створити свою функцію в Python.

generate_list — ім’я нашої функції (можна було назвати інакше).

У дужках ми вказуємо параметри, які можна змінювати при виклику:

prefix → початок кожного значення (наприклад "Lv_" або "Lv_Vitr_SKF").

letters → літери, які додаються після числа або після prefix.

start та end → числа від і до (можна не вказувати).

None означає "нічого не задано" (за замовчуванням).

2️⃣ result = []
Створюємо порожній список — сюди будемо додавати всі згенеровані значення.

3️⃣ if start is not None and end is not None:
Перевіряємо, чи користувач передав числа.

Якщо числа є — треба створювати комбінації "prefix + число + літера".

Якщо чисел немає — просто "prefix + літера".

4️⃣ Вкладені цикли for
for num in range(start, end + 1): — йдемо по всіх числах від start до end включно.

for letter in letters: — для кожного числа перебираємо всі літери.

result.append(f"{prefix}{num}{letter}") — створюємо текст і додаємо його в список.

f"..." — це f-рядок, зручний спосіб вставляти змінні прямо в текст.

5️⃣ else:
Якщо чисел немає (start і end не задані) — робимо простий цикл по літерах і додаємо тільки prefix + літера.

6️⃣ return result
Повертаємо готовий список з функції, щоб ми могли його зберегти у змінну і використати далі.

7️⃣ Виклик функції
python
Копіювати
Редагувати
list1 = generate_list("Lv_", "ABCDE", 49, 60)
"Lv_" → початок

"ABCDE" → літери

49, 60 → числа від 49 до 60 включно.

8️⃣ Вивід
python
Копіювати
Редагувати
print("\n".join(list1))
"\n".join(list1) → об’єднує список у один великий рядок, де кожен елемент з нового рядка.

python
Копіювати
Редагувати
print("Кількість:", len(list1))
len(list1) → показує кількість елементів у списку.


Код із поясненням
import pandas as pd
from google.colab import files


pandas – бібліотека для роботи з таблицями/CSV. Дуже зручна для обробки даних.

google.colab.files – спеціальний модуль у Colab, щоб завантажувати/вивантажувати файли між твоїм комп’ютером і Colab.

uploaded = files.upload()  # обираєш ids.csv зі свого комп’ютера


Відкриває вікно для завантаження файлу з комп’ютера в середовище Colab.

Ми завантажуємо ids.csv.

df = pd.read_csv("ids.csv", header=None, dtype=str)  # без заголовків, як рядки
ids = df[0].tolist()


pd.read_csv("ids.csv", header=None, dtype=str)

header=None → каже, що у файлі немає заголовків, це просто список.

dtype=str → зчитує все як рядки, щоб не втратилися нулі на початку (бо якби було int, то 00000065485 перетворилося б на 65485).

df[0].tolist() – беремо перший (і єдиний) стовпець і перетворюємо у список Python.

formatted_ids = [f'"{i}"' for i in ids]


Це list comprehension → короткий спосіб створення нового списку.

Для кожного i в ids додаємо лапки навколо нього.

Результат:

["00000065485", "00000071819", "00000178735", ...]

output_file = "formatted_ids.txt"
with open(output_file, "w") as f:
    f.write(",\n".join(formatted_ids))


Відкриваємо файл formatted_ids.txt для запису.

,\n".join(formatted_ids) → об’єднує список у великий рядок,
де між елементами ставиться кома + новий рядок.

Приклад результату у файлі:

"00000065485",
"00000071819",
"00000178735",
"10000192822"

files.download(output_file)


Дає посилання для скачування файлу з Colab назад на твій комп’ютер.

🔹 Підсумок

Ми використали:

pandas → щоб легко зчитати CSV навіть з тисячами рядків.

list comprehension → щоб швидко обгорнути всі значення в лапки.

join → щоб зробити акуратний текстовий формат.

google.colab.files → для зручного обміну файлами між Colab і твоїм компом.
```
